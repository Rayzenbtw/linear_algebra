–ü—Ä–∏–ø—É—Å—Ç–∏–º–æ –º–∏ –º–∞—î–º–æ –¥–≤—ñ –≥–∞–ª—É–∑—ñ –µ–∫–æ–Ω–æ–º—ñ–∫–∏ (–Ω–∞–ø—Ä–∏–∫–ª–∞–¥, –≤–∏—Ä–æ–±–Ω–∏—Ü—Ç–≤–æ —Å—Ç–∞–ª—ñ ($B_1$), —Ç–∞ –≤–∏—Ä–æ–±–Ω–∏—Ü—Ç–≤–æ –∞–≤—Ç–æ–º–æ–±—ñ–ª–µ–π ($B_2$)). –ö–æ–∂–Ω–∞ –∑ —Ü–∏—Ö –≥–∞–ª—É–∑–µ–π $B_i$ –≤–∏—Ä–æ–±–ª—è—î –ø—Ä–æ–¥—É–∫—Ü—ñ—é, —è–∫—É —Å–ø–æ–∂–∏–≤–∞—é—Ç—å: –≤–æ–Ω–∞ —Å–∞–º–∞ ($d_{ii}$), –¥—Ä—É–≥–∞ –∑ —Ä–æ–∑–≥–ª—è–¥–∞–Ω–∏—Ö –≥–∞–ª—É–∑–µ–π ($d_{ij}$), –∞ —Ç–∞–∫–æ–∂ —ñ–Ω—à—ñ –≥–∞–ª—É–∑—ñ –µ–∫–æ–Ω–æ–º—ñ–∫–∏ ($d_{i0}$). –ù–∞–ø—Ä–∏–∫–ª–∞–¥, –∑–∞ –æ–∑–Ω–∞—á–µ–Ω–∏–º–∏ –≥–∞–ª—É–∑—è–º–∏ —É –¥–µ—è–∫–æ–º—É —Ä–æ—Ü—ñ —Å–ø–æ—Å—Ç–µ—Ä—ñ–≥–∞—î—Ç—å—Å—è —Å–∏—Ç—É–∞—Ü—ñ—è, –Ω–∞–≤–µ–¥–µ–Ω–∞ –≤ —Ç–∞–±–ª–∏—Ü—ñ –Ω–∏–∂—á–µ. –î–æ —Ä–µ—á—ñ, —Ü—ñ –¥–∞–Ω—ñ –Ω–∞–≤–µ–¥–µ–Ω–æ —â–æ–¥–æ –µ–∫–æ–Ω–æ–º—ñ–∫–∏ –°–®–ê –∑–∞ 1958 —Ä—ñ–∫ —É –º–ª–Ω.–¥–æ–ª. –∑—ñ —Å—Ç–∞—Ç—Ç—ñ –Ω–æ–±–µ–ª—ñ–≤—Å—å–∫–æ–≥–æ –ª–∞—É—Ä–µ–∞—Ç–∞ –∑ –µ–∫–æ–Ω–æ–º—ñ–∫–∏ –í–∞—Å–∏–ª—è –õ–µ–æ–Ω—Ç—å—î–≤–∞: Wassily W. Leontief, The Structure of the U.S. Economy, Scientific
American, volume 212 number 4 (Apr. 1965), p. 2. –Ü–¥–µ—è –∞–Ω–∞–ª—ñ–∑—É, –≤—ñ–¥–æ–º–æ–≥–æ —Ç–∞–∫–æ–∂ —è–∫ input-output analysis, —Ç–∞–∫–æ–∂ –Ω–∞–ª–µ–∂–∏—Ç—å –π–æ–º—É.

```python
df_economy = pd.DataFrame(columns = ['branch', 'B1_consumption', 'B2_consumption', 'other_consumption', 'total'])
df_economy['branch'] = ['B1', 'B2']
df_economy['B1_consumption'] = [5395, 48]
df_economy['B2_consumption'] = [2664, 9030]
df_economy['other_consumption'] = ['', '']
df_economy['total'] = [25448, 30346]
df_economy
```

–ó —Ü—ñ—î—ó —Ç–∞–±–ª–∏—Ü—ñ, –∑–æ–∫—Ä–µ–º–∞ –º–æ–∂–Ω–∞ –¥–æ–≤—ñ–¥–∞—Ç–∏—Å—è, —â–æ —Å—Ç–∞–ª–µ–ª–∏–≤–∞—Ä–Ω–∞ –ø—Ä–æ–º–∏—Å–ª–æ–≤—ñ—Å—Ç—å —Å–ø–æ–∂–∏–≤–∞—î –≤–ª–∞—Å–Ω–æ—ó –ø—Ä–æ–¥—É–∫—Ü—ñ—ó –Ω–∞ —Å—É–º—É $d_{11} = 5395$ –º–ª–Ω. –¥–æ–ª., –∞ –∞–≤—Ç–æ–º–æ–±—ñ–ª–µ–±—É–¥—É–≤–∞–Ω–Ω—è —Å–ø–æ–∂–∏–≤–∞—î —Å—Ç–∞–ª—ñ –Ω–∞ —Å—É–º—É $d_{12} = 2664$ –º–ª–Ω. –¥–æ–ª. –¢–∞–∫–æ–∂, –∑–Ω–∞—é—á–∏ –∑–∞–≥–∞–ª—å–Ω–∏–π –æ–±—Å—è–≥ –≤–∏—Ä–æ–±–Ω–∏—Ü—Ç–≤–∞ —Å—Ç–∞–ª—ñ $D_1=25448$, –º–æ–∂–Ω–∞  —Ä–æ–∑—Ä–∞—Ö—É–≤–∞—Ç–∏ —Å–∫—ñ–ª—å–∫–∏ —Å—Ç–∞–ª—ñ —Å–ø–æ–∂–∏–≤–∞—é—Ç—å —ñ–Ω—à—ñ –≥–∞–ª—É–∑—ñ –µ–∫–æ–Ω–æ–º—ñ–∫–∏ (–∑–æ–∫—Ä–µ–º–∞ –π —ñ–¥–µ –Ω–∞ –µ–∫—Å–ø–æ—Ä—Ç, –∞–±–æ –∑–∞–ª–∏—à–∞—î—Ç—å—Å—è –Ω–∞ —Å–∫–ª–∞–¥–∞—Ö): $d_{10}=D_1-d_{11}-d_{12}$. –ó—Ä–æ–±—ñ—Ç—å —Ü–µ –¥–ª—è –æ–±–æ—Ö –≥–∞–ª—É–∑–µ–π $B_1$, $B_2$ –π –∑–∞–ø–∏—à—ñ—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏ —É —Å—Ç–æ–≤–ø—á–∏–∫ 'other_consumption'.   

```python
#write your code here
df_economy['other_consumption'] = df_economy['total'] - df_economy['B1_consumption'] - df_economy['B2_consumption']
df_economy
```

–ù–µ—Ö–∞–π –º–∏ –º–∞—î–º–æ –ø—Ä–æ–≥–Ω–æ–∑ —â–æ–¥–æ —Ç–æ–≥–æ, —è–∫ –∑–º—ñ–Ω–∏—Ç—å—Å—è —Å–ø–æ–∂–∏–≤–∞–Ω–Ω—è —Å—Ç–∞–ª—ñ —Ç–∞ –∞–≤—Ç–æ –∑–æ–≤–Ω—ñ—à–Ω—ñ–º–∏ —Å–ø–æ–∂–∏–≤–∞—á–∞–º–∏ ('other\_consumption') –Ω–∞ –Ω–∞—Å—Ç—É–ø–Ω–∏–π —Ä—ñ–∫:

$d'_{10} =d_{10} +\Delta _{1}$, $d'_{20} =d_{20} +\Delta _{2}$, $\Delta _{1} =200$, $\Delta _{2} =-25$. –¢–æ–±—Ç–æ —Å–ø–æ–∂–∏–≤–∞–Ω–Ω—è —Å—Ç–∞–ª—ñ –∑–æ–≤–Ω—ñ—à–Ω—ñ–º–∏ —Å–ø–æ–∂–∏–≤–∞—á–∞–º–∏ –∑—Ä–æ—Å—Ç–µ –Ω–∞ 200 –º–ª–Ω. –¥–æ–ª., –∞ —Å–ø–æ–∂–∏–≤–∞–Ω–Ω—è –∞–≤—Ç–æ –∑–º–µ–Ω—à–∏—Ç—å—Å—è –Ω–∞ 25 –º–ª–Ω. –¥–æ–ª.



–ù–∞—à–∞ –º–µ—Ç–∞ —Ä–æ–∑—Ä–∞—Ö—É–≤–∞—Ç–∏, —Å–∫—ñ–ª—å–∫–∏ —Å—Ç–∞–ª—ñ —Ç–∞ –∞–≤—Ç–æ –º–∞—î –≤–∏—Ä–æ–±–ª—è—Ç–∏—Å—è –¥–ª—è —Ç–æ–≥–æ, —â–æ–± –∑–∞–¥–æ–≤–æ–ª—å–Ω–∏—Ç–∏ —É—Å—ñ –ø,–æ—Ç—Ä–µ–±–∏ –∑–≤–∞–∂–∞—é—á–∏ –Ω–∞ –ø–µ—Ä–µ–¥–±–∞—á—É–≤–∞–Ω—ñ –∑–º—ñ–Ω–∏ —É —Å–ø–æ–∂–∏–≤–∞–Ω–Ω—ñ. 



–î–ª—è —Ü—å–æ–≥–æ –º–∏ –∑–∞–ø–∏—à–µ–º–æ —É–º–æ–≤–∏, —â–æ –ø–æ–≤'—è–∑—É—é—Ç—å —É—Å—ñ –Ω–∞–≤–µ–¥–µ–Ω—ñ —Ç–∞ —à—É–∫–∞–Ω—ñ –≤–µ–ª–∏—á–∏–Ω–∏.



–°–ø–æ–∂–∏–≤–∞–Ω–Ω—è —Ü—å–æ–≥–æ —Ä–æ–∫—É (–∑–Ω–∞—á–µ–Ω–Ω—è –≤—Å—ñ—Ö –≤–µ–ª–∏—á–∏–Ω –Ω–∞–º –≤—ñ–¥–æ–º—ñ –∑ —Ç–∞–±–ª–∏—Ü—ñ) –∑–∞–¥–∞—î—Ç—å—Å—è —Å–∏—Å—Ç–µ–º–æ—é —É–º–æ–≤:



${\displaystyle \begin{cases}
D_{1} =d_{11} +d_{12} +d_{10} ;\\
D_{2} =d_{21} +d_{22} +d_{20}
\end{cases}}$



–°–ø–æ–∂–∏–≤–∞–Ω–Ω—è –Ω–∞—Å—Ç—É–ø–Ω–æ–≥–æ —Ä–æ–∫—É:



${\displaystyle \begin{cases}
D'_{1} =d'_{11} +d'_{12} +d'_{10} ;\\
D'_{2} =d'_{21} +d'_{22} +d'_{20}
\end{cases}}$



–£ –¥—Ä—É–≥—ñ–π —Å–∏—Å—Ç–µ–º—ñ —É–º–æ–≤ –Ω–∞–º –≤—ñ–¥–æ–º—ñ –ª–∏—à–µ –∑–Ω–∞—á–µ–Ω–Ω—è –≤–µ–ª–∏—á–∏–Ω $d'_{10} ,d'_{20}$. –î–ª—è —Ç–æ–≥–æ, —â–æ–± –∑–Ω–∞–π—Ç–∏ $D'_{1} ,D'_{2}$ –º–∏ –±—É–¥–µ–º–æ –¥–æ–¥–µ—Ä–∂—É–≤–∞—Ç–∏—Å—è –ø—Ä–∏–ø—É—â–µ–Ω–Ω—è, —â–æ —á–∞—Å—Ç–∫–∞ —Å–ø–æ–∂–∏–≤–∞–Ω–Ω—è –≥–∞–ª—É–∑–∑—é –≤–ª–∞—Å–Ω–æ—ó –ø—Ä–æ–¥—É–∫—Ü—ñ—ó —Ç–∞ –ø—Ä–æ–¥—É–∫—Ü—ñ—ó –¥—Ä—É–≥–æ—ó –∑ —Ä–æ–∑–≥–ª—è–¥–∞–Ω–∏—Ö –≥–∞–ª—É–∑–µ–π –∑–∞–ª–∏—à–∞—î—Ç—å—Å—è –Ω–µ–∑–º—ñ–Ω–Ω–æ—é –∑ —Ä–æ–∫—É –≤ —Ä—ñ–∫, —Ç–æ–±—Ç–æ: $\dfrac{d'_{ij}}{D'_{i}} =\dfrac{d_{ij}}{D_{i}}$. –ù–∞–ø—Ä–∏–∫–ª–∞–¥, —è–∫—â–æ —Ü—å–æ–≥–æ —Ä–æ–∫—É –∞–≤—Ç–æ–ø—Ä–æ–º–∏—Å–ª–æ–≤—ñ—Å—Ç—å —Å–ø–æ–∂–∏–≤–∞–ª–∞ –ø—Ä–∏–±–ª–∏–∑–Ω–æ 10\% —É—Å—ñ—î—ó —Å—Ç–∞–ª—ñ, —Ç–æ –Ω–∞—Å—Ç—É–ø–Ω–æ–≥–æ —Ä–æ–∫—É –≤–æ–Ω–∞ —Å–ø–æ–∂–∏–≤–∞—Ç–∏–º–µ —Ç–∞–∫–∏–π —Å–∞–º–∏–π –≤—ñ–¥—ñ–æ—Ç–æ–∫.



–ó –æ—Å—Ç–∞–Ω–Ω—å–æ—ó —Ä—ñ–≤–Ω–æ—Å—Ç—ñ –≤–∏—Ä–∞–∑—ñ—Ç—å –≤–µ–ª–∏—á–∏–Ω–∏ $d'_{ij}$ —Ç–∞ –ø—ñ–¥—Å—Ç–∞–≤—Ç–µ –≤ –¥—Ä—É–≥—É —Å–∏—Å—Ç–µ–º—É. –¢–∞–∫–æ–∂ –ø—ñ–¥—Å—Ç–∞–≤—Ç–µ –∑–Ω–∞—á–µ–Ω–Ω—è $d' _{10} ,d'_{20}$. –ú–∞—î –∑–∞–ª–∏—à–∏—Ç–∏—Å—è —Å–∏—Å—Ç–µ–º–∞ —Ä—ñ–≤–Ω—è–Ω—å –∑ –¥–≤–æ–º–∞ –Ω–µ–≤—ñ–¥–æ–º–∏–º–∏ $D'_{1} ,D'_{2}$.

–ó–∞–ø–∏—à—ñ—Ç—å –º–∞—Ç—Ä–∏—Ü—é —Å–∏—Å—Ç–µ–º–∏, —ó—ó –ø—Ä–∞–≤—É —á–∞—Å—Ç–∏–Ω—É, —Ç–∞ —Ä–æ–∑–≤'—è–∂—ñ—Ç—å —Å–∏—Å—Ç–µ–º—É, —â–æ–± –∑–Ω–∞–π—Ç–∏  $D'_1, D'_2$.



‚úîÔ∏è 
## –£ –º–∞—Ç—Ä–∏—á–Ω–æ–º—É –≤–∏–≥–ª—è–¥—ñ:

–ú–∞—î–º–æ —Å–∏—Å—Ç–µ–º—É —Ä—ñ–≤–Ω—è–Ω—å:

$$ A \cdot X = B $$

–¥–µ:
- \( A \) ‚Äî –º–∞—Ç—Ä–∏—Ü—è –∫–æ–µ—Ñ—ñ—Ü—ñ—î–Ω—Ç—ñ–≤,
- \( X \) ‚Äî –≤–µ–∫—Ç–æ—Ä –Ω–µ–≤—ñ–¥–æ–º–∏—Ö,
- \( B \) ‚Äî –≤–µ–∫—Ç–æ—Ä –ø—Ä–∞–≤–∏—Ö —á–∞—Å—Ç–∏–Ω.

### –ú–∞—Ç—Ä–∏—Ü—ñ:

$$
A = \begin{bmatrix}
\frac{d_{10}}{D_1} & 0 \\
0 & \frac{d_{20}}{D_2} 
\end{bmatrix}  
$$

$$
X = \begin{bmatrix} D'_1 \\ D'_2 \end{bmatrix}  
$$

$$
B = \begin{bmatrix} d'_{10} \\ d'_{20} \end{bmatrix}  
$$

### –û–±–µ—Ä–Ω–µ–Ω–∞ –º–∞—Ç—Ä–∏—Ü—è \( A^{-1} \):

–û–±–µ—Ä–Ω–µ–Ω—É –º–∞—Ç—Ä–∏—Ü—é \( A^{-1} \) –º–æ–∂–Ω–∞ –∑–Ω–∞–π—Ç–∏ –∑–∞ –ø—Ä–∞–≤–∏–ª–æ–º –æ–±–µ—Ä–Ω–µ–Ω–Ω—è –¥—ñ–∞–≥–æ–Ω–∞–ª—å–Ω–æ—ó –º–∞—Ç—Ä–∏—Ü—ñ:

$$
A^{-1} = \begin{bmatrix}
\frac{D_1}{d_{10}} & 0 \\
0 & \frac{D_2}{d_{20}} 
\end{bmatrix}  
$$

### –†–æ–∑–≤‚Äô—è–∑–æ–∫:

–ó–Ω–∞–π–¥–µ–º–æ \( X \):

$$
X = A^{-1} \cdot B 
$$
`
$$
\begin{bmatrix} D'_1 \\ D'_2 \end{bmatrix} =
\begin{bmatrix}
\frac{D_1}{d_{10}} & 0 \\
0 & \frac{D_2}{d_{20}} 
\end{bmatrix}
\cdot
\begin{bmatrix} d'_{10} \\ d'_{20} \end{bmatrix}  
$$

–í–∏–∫–æ–Ω–∞—î–º–æ –º–Ω–æ–∂–µ–Ω–Ω—è –º–∞—Ç—Ä–∏—Ü—å:

$$
D'_1 = \frac{D_1}{d_{10}} \cdot d'_{10} = D_1 \cdot \frac{d'_{10}}{d_{10}}
$$

$$
D'_2 = \frac{D_2}{d_{20}} \cdot d'_{20} = D_2 \cdot \frac{d'_{20}}{d_{20}}
$$

–¢–∞–∫–∏–º —á–∏–Ω–æ–º, –º–∏ –æ—Ç—Ä–∏–º–∞–ª–∏ –∑–Ω–∞—á–µ–Ω–Ω—è D'_1 —Ç–∞ D'_2.

### –í–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è –∑ –ø–µ—Ä—à–æ—ó —á–∞—Å—Ç–∏–Ω–∏ –∑–∞–≤–¥–∞–Ω–Ω—è:

$$
D_1 = 25448, \quad D_2 = 30346
$$

$$
d_{11} = 5395, \quad d_{12} = 2664
$$

$$
d_{21} = 48, \quad d_{22} = 9030
$$

$$
d_{10} = 17389, \quad d_{20} = 21268
$$
### –ü—ñ–¥—Å—Ç–∞–≤–∏–º–æ –Ω–∞—à—ñ –∑–Ω–∞—á–µ–Ω–Ω—è:
$$
D'_1 = \frac{25448}{17389} \times 17589 = 25740.691
$$

$$
D'_2 = \frac{30346}{21268} \times 21243 = 30310.329
$$



–ó—Ä–æ–±—ñ—Ç—å —Å–ø–æ—Å—Ç–µ—Ä–µ–∂–µ–Ω–Ω—è: —á–∏ –≤—ñ—Ä–Ω–æ, —â–æ –≤–∏—Ä–æ–±–Ω–∏—Ü—Ç–≤–æ —Å—Ç–∞–ª—ñ ($D'_1$) –∑—Ä–æ—Å–ª–æ –Ω–∞ –≤–µ–ª–∏—á–∏–Ω—É $\Delta_1$, –∞ –≤–∏—Ä–æ–±–Ω–∏—Ü—Ç–≤–æ –∞–≤—Ç–æ ($D'_2$) –∑–º–µ–Ω—à–∏–ª–æ—Å—è –Ω–∞ –≤–µ–ª–∏—á–∏–Ω—É $\Delta_2$? –Ø–∫ –∑–º—ñ–Ω–∏–ª–∏—Å—è —Ü—ñ –≤–µ–ª–∏—á–∏–Ω–∏ –ø–æ—Ä—ñ–≤–Ω—è–Ω–æ –∑ –ø–æ–ø–µ—Ä–µ–¥–Ω—ñ–º —Ä–æ–∫–æ–º? –ß–∏ –º–æ–∂–µ—Ç–µ –í–∏ –≤–∏—Å–ª–æ–≤–∏—Ç–∏ –ø—Ä–∏–ø—É—â–µ–Ω–Ω—è, —á–æ–º—É —Ç–∞–∫ —Å—Ç–∞–ª–æ—Å—è?

‚úîÔ∏è
### –ù—ñ, –≤–æ–Ω–∏ –∑–º—ñ–Ω–∏–ª–∏—Å—è –Ω–µ –Ω–∞ –∑–Ω–∞—á–µ–Ω–Ω—è $\Delta_1$ —Ç–∞ $\Delta_2$, –≤–æ–Ω–∏ –∑–º—ñ–Ω–∏–ª–∏—Å—è –Ω–∞ —ñ–Ω—à—ñ –∑–Ω–∞—á–µ–Ω–Ω—è. –ü—Ä–∏–ø—É—â–µ–Ω–Ω—è: —á–µ—Ä–µ–∑ —Ç–µ, —â–æ —É –Ω–∞—Å —á–∞—Å—Ç–∫–∞ —Å–ø–æ–∂–∏–≤–∞–Ω–Ω—è –∑–∞–ª–∏—à–∏–ª–∞—Å—è —Ç–∞–∫–æ—é –∂, —Ç–æ –∑–Ω–∞—á–µ–Ω–Ω—è –±—É–¥—É—Ç—å –ø—Ä–æ–ø–æ—Ä—Ü—ñ–π–Ω–æ –∑–º—ñ–Ω—é–≤–∞—Ç–∏—Å—è(–∑–≥—ñ–¥–Ω–æ –∑ —Ä–æ–∑–≤'—è–∑–∫–æ–º)

2Ô∏è‚É£ –£–∑–∞–≥–∞–ª—å–Ω—ñ—Ç—å —ñ–¥–µ—é —Ç–∞–∫–æ–≥–æ –∞–Ω–∞–ª—ñ–∑—É –Ω–∞ –¥–æ–≤—ñ–ª—å–Ω—É –∫—ñ–ª—å–∫—ñ—Å—Ç—å –æ–∫—Ä–µ–º–æ —Ä–æ–∑–≥–ª—è–Ω—É—Ç–∏—Ö –≥–∞–ª—É–∑–µ–π: $B_1, B_2, \dots , B_n$ —Ç–∞ –∑–æ–≤–Ω—ñ—à–Ω—ñ—Ö —Å–ø–æ–∂–∏–≤–∞—á—ñ–≤ ('other consumption').
–í–≤–∞–∂–∞–π—Ç–µ, —â–æ –≤–∞–º –∑–∞–¥–∞–Ω–æ –≤–µ–∫—Ç–æ—Ä –∑–∞–≥–∞–ª—å–Ω–æ–≥–æ –≤–∏—Ä–æ–±–Ω–∏—Ü—Ç–≤–∞ (—ñ —Å–ø–æ–∂–∏–≤–∞–Ω–Ω—è) $D = (D_1, \dots , D_n)$, –º–∞—Ç—Ä–∏—Ü—é —Å–ø–æ–∂–∏–≤–∞–Ω–Ω—è –≥–∞–ª—É–∑—è–º–∏ $(d_{ij})$ –Ω–∞ —Ü–µ–π —Ä—ñ–∫, –∞ —Ç–∞–∫–æ–∂ –≤–µ–∫—Ç–æ—Ä –ø—Ä–æ–≥–Ω–æ–∑—ñ–≤ –∑–º—ñ–Ω —É —Å–ø–æ–∂–∏–≤–∞–Ω–Ω—ñ –∑–æ–≤–Ω—ñ—à–Ω—ñ–º–∏ —Å–ø–æ–∂–∏–≤–∞—á–∞–º–∏ –Ω–∞ –Ω–∞—Å—Ç—É–ø–Ω–∏–π —Ä—ñ–∫ $\Delta = (\Delta_1, \dots, \Delta_n)$, —Ç–∞–∫–∏–π —â–æ $d'_{i0}=d_{i0}+\Delta_i$. –ù–∞–ø–∏—à—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é, —è–∫–∞ —Ä–æ–∑—Ä–∞—Ö–æ–≤—É—î –≤–µ–∫—Ç–æ—Ä –≤–∏—Ä–æ–±–Ω–∏—Ü—Ç–≤–∞ –∑–∞ –≥–∞–ª—É–∑—è–º–∏ –Ω–∞ –Ω–∞—Å—Ç—É–ø–Ω–∏–π —Ä—ñ–∫: $D' = (D'_1, \dots , D'_n)$. 

```python
import numpy as np
import pandas as pd

def calculate_production_next_year(D, consumption_matrix, delta):
    """
    –†–æ–∑—Ä–∞—Ö–æ–≤—É—î –≤–µ–∫—Ç–æ—Ä –≤–∏—Ä–æ–±–Ω–∏—Ü—Ç–≤–∞ –Ω–∞ –Ω–∞—Å—Ç—É–ø–Ω–∏–π —Ä—ñ–∫ –¥–ª—è n –≥–∞–ª—É–∑–µ–π.
    
    Parameters:
        D (list or np.array): –í–µ–∫—Ç–æ—Ä –∑–∞–≥–∞–ª—å–Ω–æ–≥–æ –≤–∏—Ä–æ–±–Ω–∏—Ü—Ç–≤–∞ (D_1, ..., D_n) —Ü—å–æ–≥–æ —Ä–æ–∫—É.
        consumption_matrix (np.array): –ú–∞—Ç—Ä–∏—Ü—è —Å–ø–æ–∂–∏–≤–∞–Ω–Ω—è (d_ij), —Ä–æ–∑–º—ñ—Ä–æ–º n x n.
        delta (list or np.array): –í–µ–∫—Ç–æ—Ä –ø—Ä–æ–≥–Ω–æ–∑–æ–≤–∞–Ω–∏—Ö –∑–º—ñ–Ω –∑–æ–≤–Ω—ñ—à–Ω—å–æ–≥–æ —Å–ø–æ–∂–∏–≤–∞–Ω–Ω—è (Œî_1, ..., Œî_n).
    
    Returns:
        np.array: –í–µ–∫—Ç–æ—Ä –≤–∏—Ä–æ–±–Ω–∏—Ü—Ç–≤–∞ –Ω–∞ –Ω–∞—Å—Ç—É–ø–Ω–∏–π —Ä—ñ–∫ (D'_1, ..., D'_n).
    """
    n = len(D)
    D = np.array(D)
    delta = np.array(delta)
    consumption_matrix = np.array(consumption_matrix)
    
    # –û–±—á–∏—Å–ª—é—î–º–æ –∑–æ–≤–Ω—ñ—à–Ω—î —Å–ø–æ–∂–∏–≤–∞–Ω–Ω—è —Ü—å–æ–≥–æ —Ä–æ–∫—É
    external_consumption = D - np.sum(consumption_matrix, axis=1)
    
    # –ü—Ä–æ–≥–Ω–æ–∑–æ–≤–∞–Ω–µ –∑–æ–≤–Ω—ñ—à–Ω—î —Å–ø–æ–∂–∏–≤–∞–Ω–Ω—è –Ω–∞—Å—Ç—É–ø–Ω–æ–≥–æ —Ä–æ–∫—É
    external_consumption_next_year = external_consumption + delta
    
    # –û–±—á–∏—Å–ª—é—î–º–æ –∫–æ–µ—Ñ—ñ—Ü—ñ—î–Ω—Ç–∏ a_i
    coefficients = 1 - np.sum(consumption_matrix / D[:, None], axis=1)
    
    # –û–±—á–∏—Å–ª—é—î–º–æ –≤–∏—Ä–æ–±–Ω–∏—Ü—Ç–≤–æ –Ω–∞ –Ω–∞—Å—Ç—É–ø–Ω–∏–π —Ä—ñ–∫
    D_next_year = external_consumption_next_year / coefficients
    
    return D_next_year

# –¢–µ—Å—Ç–æ–≤–∏–π –ø—Ä–∏–∫–ª–∞–¥
if __name__ == "__main__":
    # –í—Ö—ñ–¥–Ω—ñ –¥–∞–Ω—ñ –∑ —Ç–∞–±–ª–∏—Ü—ñ
    D = [800, 300, 500]  # –í–µ–∫—Ç–æ—Ä –∑–∞–≥–∞–ª—å–Ω–æ–≥–æ –≤–∏—Ä–æ–±–Ω–∏—Ü—Ç–≤–∞ —Ü—å–æ–≥–æ —Ä–æ–∫—É
    consumption_matrix = [
        [25, 50, 100],  # –°–ø–æ–∂–∏–≤–∞–Ω–Ω—è B1
        [25, 50, 50],   # –°–ø–æ–∂–∏–≤–∞–Ω–Ω—è B2
        [15, 10, 0]     # –°–ø–æ–∂–∏–≤–∞–Ω–Ω—è B3
    ]
    delta = [625, 200, 475]  # –í–µ–∫—Ç–æ—Ä –ø—Ä–æ–≥–Ω–æ–∑–æ–≤–∞–Ω–∏—Ö –∑–º—ñ–Ω –∑–æ–≤–Ω—ñ—à–Ω—å–æ–≥–æ —Å–ø–æ–∂–∏–≤–∞–Ω–Ω—è

    # –í–∏–∫–ª–∏–∫ —Ñ—É–Ω–∫—Ü—ñ—ó
    D_next_year = calculate_production_next_year(D, consumption_matrix, delta)

    # –§–æ—Ä–º—É—î–º–æ —Ç–∞–±–ª–∏—Ü—é –¥–ª—è –≤—ñ–¥–æ–±—Ä–∞–∂–µ–Ω–Ω—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ñ–≤
    df_result = pd.DataFrame({
        'Branch': ['B1', 'B2', 'B3'],
        'Current Production (D)': D,
        'Consumption (B1)': consumption_matrix[0],
        'Consumption (B2)': consumption_matrix[1],
        'Consumption (B3)': consumption_matrix[2],
        'Delta': delta,
        'New Production (D\')': D_next_year
    })

    # –í–∏–≤–æ–¥–∏–º–æ —Ç–∞–±–ª–∏—Ü—é
    print("–¢–∞–±–ª–∏—Ü—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ñ–≤:")
    print(df_result)
```
