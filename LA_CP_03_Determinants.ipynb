```python
#write your code here
import numpy as np
from scipy import linalg

def chio_determinant(matrix):
    """
    –û–±—á–∏—Å–ª—é—î –≤–∏–∑–Ω–∞—á–Ω–∏–∫ –º–∞—Ç—Ä–∏—Ü—ñ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é "—Ç—Ä—é–∫—É –ß—ñ–æ".
    
    –ü–∞—Ä–∞–º–µ—Ç—Ä–∏:
        matrix (np.ndarray): –ö–≤–∞–¥—Ä–∞—Ç–Ω–∞ –º–∞—Ç—Ä–∏—Ü—è –ø–æ—Ä—è–¥–∫—É n.
        
    –ü–æ–≤–µ—Ä—Ç–∞—î:
        float: –í–∏–∑–Ω–∞—á–Ω–∏–∫ –º–∞—Ç—Ä–∏—Ü—ñ.
    """
    # –ü–µ—Ä–µ–≤—ñ—Ä–∫–∞, —á–∏ –º–∞—Ç—Ä–∏—Ü—è –∫–≤–∞–¥—Ä–∞—Ç–Ω–∞
    if matrix.shape[0] != matrix.shape[1]:
        raise ValueError("–ú–∞—Ç—Ä–∏—Ü—è –ø–æ–≤–∏–Ω–Ω–∞ –±—É—Ç–∏ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ—é.")
    
    n = matrix.shape[0]
    
    # –ë–∞–∑–æ–≤–∏–π –≤–∏–ø–∞–¥–æ–∫: –º–∞—Ç—Ä–∏—Ü—è 1x1
    if n == 1:
        return matrix[0, 0]
    
    # –ë–∞–∑–æ–≤–∏–π –≤–∏–ø–∞–¥–æ–∫: –º–∞—Ç—Ä–∏—Ü—è 2x2
    if n == 2:
        return matrix[0, 0] * matrix[1, 1] - matrix[0, 1] * matrix[1, 0]
    
    # –Ø–∫—â–æ a11 == 0, –¥–æ–¥–∞—î–º–æ —ñ–Ω—à–∏–π —Ä—è–¥–æ–∫ –¥–æ –ø–µ—Ä—à–æ–≥–æ
    if matrix[0, 0] == 0:
        # –®—É–∫–∞—î–º–æ –Ω–µ–Ω—É–ª—å–æ–≤–∏–π –µ–ª–µ–º–µ–Ω—Ç —É –ø–µ—Ä—à–æ–º—É —Å—Ç–æ–≤–ø—Ü—ñ
        for i in range(1, n):
            if matrix[i, 0] != 0:
                # –î–æ–¥–∞—î–º–æ i-–π —Ä—è–¥–æ–∫ –¥–æ –ø–µ—Ä—à–æ–≥–æ
                matrix[0] += matrix[i]
                break
        else:
            # –Ø–∫—â–æ –≤—Å—ñ –µ–ª–µ–º–µ–Ω—Ç–∏ –ø–µ—Ä—à–æ–≥–æ —Å—Ç–æ–≤–ø—Ü—è –Ω—É–ª—å–æ–≤—ñ
            return 0.0
    
    # –§–æ—Ä–º—É—î–º–æ –Ω–æ–≤—É –º–∞—Ç—Ä–∏—Ü—é D –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –º—ñ–Ω–æ—Ä—ñ–≤ –¥—Ä—É–≥–æ–≥–æ –ø–æ—Ä—è–¥–∫—É
    D = np.zeros((n-1, n-1))
    for i in range(1, n):
        for j in range(1, n):
            D[i-1, j-1] = matrix[0, 0] * matrix[i, j] - matrix[0, j] * matrix[i, 0]
    
    # –†–µ–∫—É—Ä—Å–∏–≤–Ω–æ –æ–±—á–∏—Å–ª—é—î–º–æ –≤–∏–∑–Ω–∞—á–Ω–∏–∫ –º–∞—Ç—Ä–∏—Ü—ñ D
    det_D = chio_determinant(D)
    
    # –ü–æ–≤–µ—Ä—Ç–∞—î–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç, –≤—Ä–∞—Ö–æ–≤—É—é—á–∏ –º–Ω–æ–∂–Ω–∏–∫
    return det_D / matrix[0, 0]**(n-2)

# –¢–µ—Å—Ç—É–≤–∞–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ—ó
if __name__ == "__main__":
    # –ü—Ä–∏–∫–ª–∞–¥ –º–∞—Ç—Ä–∏—Ü—ñ
    A = np.array([
        [0, 1, 1],
        [3, 4, -1],
        [1, 5, 1]
    ])
    
    # –û–±—á–∏—Å–ª–µ–Ω–Ω—è –≤–∏–∑–Ω–∞—á–Ω–∏–∫–∞ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é —Ç—Ä—é–∫—É –ß—ñ–æ
    det_chio = chio_determinant(A)
    print(f"–í–∏–∑–Ω–∞—á–Ω–∏–∫ (–ß—ñ–æ): {det_chio}")
    
    # –û–±—á–∏—Å–ª–µ–Ω–Ω—è –≤–∏–∑–Ω–∞—á–Ω–∏–∫–∞ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é scipy.linalg.det (–º–µ—Ç–æ–¥ –ì–∞—É—Å—Å–∞)
    det_scipy = linalg.det(A)
    print(f"–í–∏–∑–Ω–∞—á–Ω–∏–∫ (Scipy): {det_scipy}")
```

## 1. –¢–µ–æ—Ä–µ—Ç–∏—á–Ω–∞ –æ—Ü—ñ–Ω–∫–∞ —Å–∫–ª–∞–¥–Ω–æ—Å—Ç—ñ

### **"–¢—Ä—é–∫ –ß—ñ–æ":**
- –ù–∞ –∫–æ–∂–Ω–æ–º—É –∫—Ä–æ—Ü—ñ –º–∏ –∑–º–µ–Ω—à—É—î–º–æ —Ä–æ–∑–º—ñ—Ä –º–∞—Ç—Ä–∏—Ü—ñ –Ω–∞ 1, –∞–ª–µ –¥–ª—è —Ñ–æ—Ä–º—É–≤–∞–Ω–Ω—è –º–∞—Ç—Ä–∏—Ü—ñ $ D $ –ø–æ—Ç—Ä—ñ–±–Ω–æ –æ–±—á–∏—Å–ª–∏—Ç–∏ $ (n-1)^2 $ –º—ñ–Ω–æ—Ä—ñ–≤ –¥—Ä—É–≥–æ–≥–æ –ø–æ—Ä—è–¥–∫—É.
- –°–∫–ª–∞–¥–Ω—ñ—Å—Ç—å: $ O(n^4) $ (—á–µ—Ä–µ–∑ —Ä–µ–∫—É—Ä—Å–∏–≤–Ω—ñ –æ–±—á–∏—Å–ª–µ–Ω–Ω—è).

### **–ú–µ—Ç–æ–¥ –ì–∞—É—Å—Å–∞:**
- –ú–µ—Ç–æ–¥ –ì–∞—É—Å—Å–∞ –∑–≤–æ–¥–∏—Ç—å –º–∞—Ç—Ä–∏—Ü—é –¥–æ —Ç—Ä–∏–∫—É—Ç–Ω–æ–≥–æ –≤–∏–≥–ª—è–¥—É –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –æ–ø–µ—Ä–∞—Ü—ñ–π –Ω–∞–¥ —Ä—è–¥–∫–∞–º–∏.
- –°–∫–ª–∞–¥–Ω—ñ—Å—Ç—å: $ O(n^3) $.

### **–ü—Ä–∞–≤–∏–ª–æ –°–∞—Ä—Ä—é—Å–∞:**
- –ó–∞—Å—Ç–æ—Å–æ–≤—É—î—Ç—å—Å—è –ª–∏—à–µ –¥–ª—è –º–∞—Ç—Ä–∏—Ü—å $ 3 \times 3 $.
